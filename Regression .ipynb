```python
# Importing the necessary modules
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
# Loading the data frame in
original_df = pd.read_csv('https://static.bc-edx.com/ai/ail-v-1-0/m12/lesson_3/datasets/garments-worker-productivity.csv')
original_df = original_df.drop(columns=['wip', 'day'])
original_df.dropna()

original_df.info()
```

```python
# Cleaning the data
X = original_df.drop(columns=['actual_productivity'])

X = pd.get_dummies(X, dtype='int',columns=['department', 'date', 'quarter', 'team'])

y = original_df['actual_productivity'].values.reshape(-1,1)

X
```

```python
# Splitting data set into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
# Loading linear regression
lr1 = LinearRegression()

lr1.fit(X_train, y_train)

lr1_train_score = lr1.score(X_test, y_test)

lr1_train_score
```

```python
lr = sm.OLS(y_train, X_train).fit()

p_values = lr.pvalues.sort_values()

select_cols = p_values.loc[p_values < 0.05]

select_cols
```
